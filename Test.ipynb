Тестовое задание Data Scientist

Задача:
Дать рекомендации и заключения по оптимизации воронки.
Оптимизация воронки - увеличение прибыли за счет большего конверта или уменьшение костов за
счет более эффективной закупки.

Данные:
https://drive.google.com/open?id=1sHAqb4sogh45V-QDCR-pNr_3boZvcgs-

Описание полей:
    DATE_OF_EVENT - день клика, заплатить он мог мозже, но все равно тригер реги упадет сюда
    EVENT_ID - уникальный ID клика, каждый пользователь может сделать неограниченное кол-во
кликов
    USER_ID - уникальный ID пользователя, каждый пользователь может сделать неограниченное кол-
во кликов
    WAY_ID - канал\сплит
    CAMPAIGN_ID - группа баннеров, кампания, внутренний идентификатор
    PLACEMENT_ID - баннер, на который кликнул пользователь.
    INT_CAMPAIGN_ID - ID кампании
    MEDIUM_ID - промежуточный баннер с вопросами (типа какой цвет волос нарвится, в каком городе
живете, и тд.). Его могут не пройти, тогда будет отвал.
    PRODUCT_ID - заход на регистрационную форму ID продукта. Тут может заполнить форму и
перейти в событию регистрации. Или уйти.
    REG_TRIG - ивент регистрации, каждый пользователь может зарегистрироваться неограниченном
кол-ве продуктов. Но только 1 раз на каждом.
    PMNT_TRIG - тригер платежа связки "USER_ID"-"PRODUCT_ID"-"REG_TRIG". Т.е данный юзер
заплатил на данном продукте после реги. Платеж без невозможен.

# Беглый анализ данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv('./Downloads/testovoe/1.csv',parse_dates=['DATE_OF_EVENT'], index_col='DATE_OF_EVENT')
```

```python
#df2=pd.read_csv('./Downloads/testovoe/2.csv',parse_dates=['DATE_OF_EVENT'], index_col='DATE_OF_EVENT')
```

```python
df1
```

```python
df1=df1.drop_duplicates()
```

```python
df1.isnull().sum()
```

```python
WAY_ID=df1.groupby(['WAY_ID','PMNT_TRIG'],as_index=False).count().pivot_table(values='EVENT_ID', index=['WAY_ID'],columns=['PMNT_TRIG'], aggfunc=sum,fill_value=0)
```

```python
#каналы в абсолютных значения которые принесли прибыли(см.столбец 1)
WAY_ID
```

```python
#канал\сплит в долях.
WAY_ID.apply(lambda x: x/x.sum(),axis=1).reset_index().sort_values(1, ascending=False)
```

```python
gr_banner=df1.groupby(['CAMPAIGN_ID','PMNT_TRIG'],as_index=False).count().pivot_table(values='EVENT_ID', index=['CAMPAIGN_ID'],columns=['PMNT_TRIG'], aggfunc=sum,fill_value=0)
```

```python
#gr_banner-группа баннеров,которые не принесли прибыли, претендеты на удаление
gr_banner[gr_banner[1]==0].sort_values(0)
```

```python
#gr_banner-группа баннеров,которые принесли прибыли в абсол.ед.
gr_banner[gr_banner[1]!=0].sort_values(1,ascending=False)
```

```python
#gr_banner-группа баннеров,которые принесли прибыли в долях.
gr_banner[gr_banner[1]!=0].apply(lambda x: x/x.sum(),axis=1).sort_values(1, ascending=False)
```

```python
PLACEMENT_banner=df1.groupby(['PLACEMENT_ID','PMNT_TRIG'],as_index=False).count().pivot_table(values='EVENT_ID', index=['PLACEMENT_ID'],columns=['PMNT_TRIG'], aggfunc=sum,fill_value=0)
```

```python
#banner-баннеры,которые не принесли прибыли, претендеты на удаление
PLACEMENT_banner[PLACEMENT_banner[1]==0].sort_values(0)
```

```python
#banner-баннеры,которые приносят прибыли
PLACEMENT_banner[PLACEMENT_banner[1]!=0].sort_values(1,ascending=False)
```

```python
MEDIUM_ID=df1.groupby(['MEDIUM_ID','PMNT_TRIG'],as_index=False).count().pivot_table(values='EVENT_ID', index=['MEDIUM_ID'],columns=['PMNT_TRIG'], aggfunc=sum,fill_value=0)
```

```python
#опросники,которые не приносят прибыли
MEDIUM_ID[MEDIUM_ID[1]==0].sort_values(0,ascending=False)
```

```python
#опросники,которые приносят прибыли
MEDIUM_ID[MEDIUM_ID[1]!=0].sort_values(1,ascending=False)
```

```python
#ID кампании после которых отваливаются пользователи 
df1[df1.MEDIUM_ID.isna()]['INT_CAMPAIGN_ID'].value_counts()
```

```python
#опросники после которых отваливаются пользователи и отказываются выбирать продукты
df1[df1.PRODUCT_ID.isna()]['MEDIUM_ID'].value_counts()
```

```python
PRODUCT_ID=df1.groupby(['PRODUCT_ID','PMNT_TRIG'],as_index=False).count().pivot_table(values='EVENT_ID', index=['PRODUCT_ID'],columns=['PMNT_TRIG'], aggfunc=sum,fill_value=0)
```

```python
#Продукты которые не покупают
PRODUCT_ID[PRODUCT_ID[1]==0].sort_values(0,ascending=False)
```

```python
#Продукты которые покупают
PRODUCT_ID[PRODUCT_ID[1]!=0].sort_values(1,ascending=False)
```

```python
df1.REG_TRIG.value_counts()
```

```python
df1.PMNT_TRIG.value_counts()
```
